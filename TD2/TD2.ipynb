# TP2 : Visualisation de données

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

## 1.1 Visualisation univariée

### Prédicteur quantitatif

```python
# Question 1 : Boite à moustaches des âges des passagers du Titanic
sns.boxplot(data = titanic, x = "age").set(title="Boite à moustaches de l'âge des personnes sur le Titanic", xlabel='Âge')
```

```python
# Question 2 : Histogramme des âges des passagers du Titanic
sns.histplot(data = titanic, x = "age", kde = True).set(title = "Histogramme des âges des passagers du Titanic", xlabel = "Âge")
# kde : affiche une courbe de densité aux données
# bins : nombre de barres
```

### Prédicteur qualitatif

```python
# Question 3
sns.countplot(data = titanic, x = "pclass").set(title = "Diagramme en barres des classes de passagers du Titanic", xlabel = "Classe", ylabel = "Nombre de passagers")
```

## 1.2 Visualisation multivariée

### Quantitatif vs quantitatif

```python
# Question 4 : Diagramme de dispersion
sns.scatterplot(data = iris, x = "sepal_length", y = "sepal_width").set(title = "Largeur du sépale des iris en fonction de la longueur", ylabel = "Largeur", xlabel = "Longueur")
```

```python
# Question 5
sns.scatterplot(data = iris, x = "sepal_length", y = "sepal_width", hue = "species").set(title = "Largeur du sépale des iris en fonction de la longueur", xlabel = "Longueur", ylabel = "Largeur")
```

```python
# Question 6
sns.scatterplot(data = iris, x = "sepal_length", y = "sepal_width", hue = "species", size = "petal_length").set(title = "Largeur du sépale des iris en fonction de la longueur", ylabel = "Largeur", xlabel = "Longueur")
```

```python
# Question 7 : diagramme de corrélation
corr = iris[["sepal_width", "sepal_length", "petal_width", "petal_length"]].corr()
sns.heatmap(corr, annot = True)
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(ncols=3, figsize=(15, 6))
for i, espece in enumerate(iris["species"].unique()):
    df = iris[iris["species"] == espece]
    corr = df[["sepal_width", "sepal_length", "petal_width", "petal_length"]].corr()
    sns.heatmap(corr, annot = True, ax = axs[i]).set(title = f"{espece}")
```

```python
# Question 8 : longueur des sépales en fonction de l'espèce

sns.barplot(data = iris, y = "sepal_width", x = "species")
```

```python
# Question 9 : boite à moustaches des longueurs sépales en fonction de l'espèce

sns.boxplot(data = iris, y = "sepal_width", x = "species").set(title = "Boite à moustaches des longueurs des sépales en fonction de l'espèce")
```

### Qualitatif vs qualitatif

```python
# Question 10 : Lieux d'embarcation en fonction de la classe

sns.countplot(data = titanic, hue = "embark_town", x = "pclass").set(title = "Lieux d'embarcation en fonction de la classe", ylabel = "Nombre de passagers", xlabel = "Classe")
```

```python
# Question 11 : Descripteur who en fonction de la classe

sns.countplot(data = titanic, hue = "who", x = "pclass").set(title = "Lieux d'embarcation en fonction de la classe", ylabel = "Nombre de passagers", xlabel = "Classe")
```

### Représentation multiple

```python
# Question 12 : Visualiser "age" en fonction de "class" avec 3 histogrammes
sns.displot(data = titanic, x = "age", kind = "hist", col = "pclass")
```

```python
# Question 13 : Visualiser "age" en fonction de class et who avec des histogrammes
sns.displot(data = titanic, x = "age", kind = "hist", col = "pclass", row = "who")
```

## Visualisation du jeu de données sy02-p2019.csv

```python
# Question 14 : Chargement du jeu de données
import pandas as pd
sy02_p2019 = pd.read_csv("data/sy02-p2019.csv", na_values=["ABS"])
type_ects = pd.CategoricalDtype(["A", "B", "C", "D", "E", "Fx", "F", "ABS"], ordered=True)
sy02_p2019["Note ECTS"] = sy02_p2019["Note ECTS"].astype(type_ects)
sy02_p2019
```

```python
# Question 14 : Type de la colonne "note median"

sy02_p2019["Note médian"].dtype
```

```python
# Question 15 : Effectifs de chaque branche

sns.countplot(data = sy02_p2019, x = "Branche").set(title = "Effectifs de chaque branche", ylabel = "Nombre d'étudiants", xlabel = "Branche")
```

```python
# Question 15 : Effectifs de chaque branche

sns.countplot(data = sy02_p2019, x = "Branche", hue = "Semestre").set(title = "Effectifs de chaque branche", ylabel = "Nombre d'étudiants", xlabel = "Branche")
```

```python
# Question 15 : Effectifs de chaque branche

sns.countplot(data = sy02_p2019, x = "Branche", hue = "Note ECTS", hue_order = ['A', 'B', 'C', 'D', 'E', 'F']).set(title = "Effectifs de chaque branche", ylabel = "Nombre d'étudiants", xlabel = "Branche")
```

```python
# Question 16 : Visualisation note médian

sns.histplot(data = sy02_p2019, x = "Note médian").set(title = "Distribution des notes du médian")
```

```python
# Question 16 : Visualisation note final

sns.histplot(data = sy02_p2019, x = "Note final").set(title = "Distribution des notes du final")
```

```python
# Question 16 : Diagramme de dispersion des notes médian / final

sns.scatterplot(data = sy02_p2019, x = "Note médian", y = "Note final").set(title = "Dispersion des notes médian / final")
```

```python
# Question 16 : jointplot

sns.jointplot(data=sy02_p2019, x="Note médian", y="Note final")
```

```python
# Question 17 : Boites à moustaches des notes de médian et final
import numpy as np
sy02_p2019_melted = pd.melt(sy02_p2019, id_vars=["Dernier diplôme obtenu", "Branche", "Semestre", "Correcteur médian", "Note TP", "Correcteur final", "Note ECTS"], var_name="Examen", value_name="Note")
sy02_p2019_melted["Examen"] = sy02_p2019_melted["Examen"].apply(lambda val : "Médian" if val == "Note médian" else "Final")
sy02_p2019_melted["Correcteur"] = sy02_p2019_melted.apply(lambda row : row["Correcteur médian"] if row["Examen"] == "Médian" else row["Correcteur final"], axis=1)
sy02_p2019_melted["Même correcteur"] = sy02_p2019_melted.apply(lambda row : "Oui" if row["Correcteur médian"] == row["Correcteur final"] else "Non", axis = 1)
sy02_p2019_melted
```

```python
sns.boxplot(data = sy02_p2019_melted, x = "Examen", y = "Note").set(title = "Boites à moustaches des notes des étudiants aux différents examens", xlabel = "Examen", ylabel = "Note obtenue")
```

```python
# Question 18 : Visualisation du nombre de copies corrigées sur tout le semestre (médian et final confondu) par correcteur en indiquant pour chaque correcteur la part des copies correspondant à un même étudiant (médian et final corrigés par ce même correcteur) et la part des copies qui ne sont pas corrigées par le même correcteur

sns.countplot(data = sy02_p2019_melted, x = "Correcteur", hue="Même correcteur").set(title = "Nombre de copies corrigées par correcteur", xlabel = "Correcteur", ylabel = "Nombre de copies")
```

```python
# Question 19 : Obtenir le même graphique que l'énoncé

sns.countplot(data = sy02_p2019_melted, x = "Correcteur", hue = "Examen").set(title = "Nombre de copies corrigées par correcteur", xlabel = "Correcteur", ylabel = "Nombre de copies")
```
